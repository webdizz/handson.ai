# Matplotlib

<a name='matplotlib'></a>

Matplotlib is a multi-platform data visualization library built on [NumPy](http://www.numpy.org/) arrays, and designed to work with the broader [SciPy](https://www.scipy.org/) stack.

One of [Matplotlib](https://matplotlib.org/)'s most important features is it's operability between different operating systems.

```python
# common header
%matplotlib inline

# import convention for matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt
```

Let's set a style that we're going to use for our visualizations.

```python
plt.style.use('seaborn-whitegrid')  # classic
```

<a name='matplotlib-plotting'></a>

### Plotting

To display something created by [Matplotlib](https://matplotlib.org/) we need to use `plt.show()` function like it is in code bellow:

```python
import numpy as np

x = np.linspace(0, 10, 150)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
```

<a name='matplotlib-subplots'></a>

### Subplots

There is object-oriented style of interface is available to perform complex plotting. 

In this style we obtain a reference to objects `Figure` and `Axes` and call their methods to perform plotting functions.

Basically on a plot bellow we do exactly the same as before but for 2 dedicated plots.

```python
fig, (plt1, plt2) = plt.subplots(2)

plt1.plot(x, np.sin(x))
plt2.plot(x, np.cos(x))

plt.show()
```

<a name='matplotlib-adjusting'></a>

### Plot Adjusting

[Matplotlib](https://matplotlib.org/) provides many ways to control the style of visualization like color, font, labels, etc.

#### Bellow we have styling of lines

```python
# specify color by name and line style
plt.plot(x, np.sin(x - 0), color='blue', linestyle='dashed')

# short color code (rgbcmyk)
plt.plot(x, np.sin(x - 1), color='g')

# Grayscale between 0 and 1
plt.plot(x, np.sin(x - 2), color='0.75')

plt.show()
```

#### Labeling of plots

```python
plt.plot(x, np.sin(x))
plt.title('Sine curve')

plt.xlabel('x')
plt.ylabel('$sin(x)$')

plt.show()
```

There is also a way to set several properties at once.

```python
ax = plt.axes()
ax.plot(x, np.cos(x))
ax.set(
    xlim=(0, 10),
    ylim=(-2, 2),
    xlabel='x',
    ylabel='$cos(x)$',
    title='A Cos(x)'
)

plt.show()
```

There is also a way to plot a points

```python
ax = plt.axes()
ax.plot(x, np.cos(x), marker='.')
ax.set(
    xlim=(0, 2),
    ylim=(-2, 2),
    xlabel='x',
    ylabel='$cos(x)$',
    title='A Cos(x)'
)

plt.show()
```

<a name='matplotlib-scatterring'></a>

### Scatterring

A second, more powerful method to visualize a data is by creating scatter plots. 

There is the `plt.scatter` function, which can be used very similarly to the `plt.plot` function.

The primary difference is that it can be used to create scatter plots where properties of each individual point can be individually controlled or mapped to data.

```python
plt.scatter(x, np.sin(x), marker='o')

plt.show()
```

Let's create a random scatter plot with colorful points.

```python
rnd = np.random.RandomState(0)
x = rnd.randn(50)
y = rnd.randn(50)

colors = rnd.rand(50)
sizes = 1000*rnd.rand(50)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')

plt.colorbar()

plt.show()
```

We can see that this scatter plot has given us the ability to simultaneously explore four different dimensions of the data.

Multicolor and multifeature scatter plots like this can be useful for both exploration and presentation of data.

However there is a question of efficiency between usage of `plt.plot` and `plt.scatter`, due to the fact that we can define attributes for each point individually it's obvious that the scatter plots are more resource intensive especially on huge datasets.
